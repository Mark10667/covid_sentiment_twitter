```python
# tokens = tokenizer.encode('it is alright', return_tensors='pt')
# result = model(tokens)
# result.logits
# int(torch.argmax(result.logits))/2-1

def BERT_score(tweet):
    tokens = tokenizer.encode(tweet, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))/2-1
```

<h1>Dataframe Load From BigQuery
<h1/>

```python
# Download GCP BigQuery table into a dataframe 

from google.cloud import bigquery

bqclient = bigquery.Client()

# Download a table.
table = bigquery.TableReference.from_string(
    "tweetanalysis-332220.geotweet.tweet"
)
rows = bqclient.list_rows(
    table,
    selected_fields=[
        bigquery.SchemaField("TWEET", "STRING"),
        bigquery.SchemaField("LOCATION", "STRING"),
        bigquery.SchemaField("DATE", "DATE"),
    ],
)
dataframe = rows.to_dataframe(
    # Optionally, explicitly request to use the BigQuery Storage API. As of
    # google-cloud-bigquery version 1.26.0 and above, the BigQuery Storage
    # API is used by default.
    create_bqstorage_client=True,
)

print(dataframe.head())
```

```python
# nltk
dataframe['POLARITY'] = dataframe['TWEET'].apply(lambda t: sid.polarity_scores(t)['compound'])
# BERT
# dataframe['POLARITY'] = dataframe['TWEET'].apply(lambda t: BERT_score(t))
```

```python
dataframe.head(10)
```

```python
# We need a function to process the dataframe from Bigsql
# 1) filter tweets that contain protection measures first 
# 2) generate each state's average sentiment on that day

def bq_df(df):
    searchfor = ['vaccin', 'Vaccin', 'mask', 'Mask', 'distanc', 'sanitizer', 'WFH']
    df = df.loc[df['TWEET'].str.contains('|'.join(searchfor))]
    
    df = df[["LOCATION", "DATE", "POLARITY"]]

    grouped_df = df.groupby(["DATE", "LOCATION"])
    mean_df = grouped_df.mean()
    mean_df = mean_df.reset_index()
    
    mean_df.rename(columns={"DATE": "date", "LOCATION": "location", "POLARITY": "polarity"}, inplace=True)
        
    return mean_df 
```

<h1>Covid Data
<h1/>

```python
# cases_avg_per_100k: seven-day rolling average per 100,000 residents

df = pd.read_csv('covid_cases.csv')
df['date']= pd.to_datetime(df['date'])
# extract rows from June 
mask = (df['date'] >= '2021-6-1') & (df['date'] <= '2021-6-30')
df_June = df.loc[mask]
df_June = df_June.sort_values(["state", "date"], ascending=True)
df_June.reset_index()
df_June.head(5)
```

```python
# create df_June_pct_ch that contains pct_ch_case_7_days

notwant2 = ['Guam', 'Northern Mariana Islands', 'Virgin Islands']
df_June_pct_ch = df_June.loc[~df_June['state'].isin(notwant2)]
df_June_pct_ch['cases_pct_ch'] = df_June_pct_ch.groupby('state')['cases_avg_per_100k'].pct_change()
df_June_pct_ch = df_June_pct_ch.reset_index()
df_June_pct_ch = df_June_pct_ch[['state', 'date', 'cases_avg_per_100k', 'cases_pct_ch']]
df_June_pct_ch['pct_ch_case_7_days'] = df_June_pct_ch.groupby('state')['cases_pct_ch'].rolling(min_periods=1, window=7).mean().reset_index(name='pct_ch_case_7_days')['pct_ch_case_7_days']
df_June_pct_ch["State_code"] = df_June_pct_ch["state"].apply(lambda x: r_dic[x])
df_June_pct_ch.head(5)
```

<h1>df_avg: 7-day average sentiment for each state 
</h1>

```python
a = bq_df(dataframe)
a['date'] = pd.to_datetime(a['date'])
df = a.groupby('location').apply(lambda x: x.set_index('date').resample('1D').first())
# average sentiment of past seven days 
df_avg = df.groupby(level=0)['polarity'].apply(lambda x: x.shift().rolling(min_periods=1,window=7).mean()).reset_index(name='Value_Average_Past_7_days')
df_avg.head(10)
```

<h1>Correlation between 7-day average cases per 100k and 7-day average sentiment 
</h1>

```python
# calculate the pearson correlation 

# align the state column of two df 
# df_June['state'] = df_June['state'].apply(lambda state : state[0]+state[1:].lower())
df_June = df_June[['state', 'date', 'cases_avg_per_100k']].reset_index()
df_avg = df_avg.rename(columns={'location': 'state'})

all_in_one = df_June.merge(df_avg, on=['state', 'date'])
corr = all_in_one.groupby('state').apply(lambda x: pearsonr(x['cases_avg_per_100k'].to_numpy(), x['Value_Average_Past_7_days'].to_numpy())[0]).reset_index()
corr = corr.rename(columns={0: 'corr'})
corr.mean()
```

<h1>Visualization: 7-day average Sentiment in June 
</h1>

```python
# df_J contains Value_Average_Past_7_days in June 
df_avg["State_code"] = df_avg["state"].apply(lambda x: r_dic[x])
notwant = ['2021-05-26', '2021-05-27', '2021-05-28', '2021-05-29', '2021-05-30', '2021-05-31']
df_J = df_avg.loc[~df_avg['date'].isin(notwant)]
df_J['Value_Average_Past_7_days'] = df_J['Value_Average_Past_7_days'].round(5)
df_J.head(5)
```

```python
# 7-day average Sentiment in June 

### colorscale:    
scl = [[0.0, '#ffffff'],[0.2, '#b4a8ce'],[0.4, '#8573a9'],
       [0.6, '#7159a3'],[0.8, '#5732a1'],[1.0, '#2c0579']] # purples

### create empty list for data object:    
data_slider = []

#### I populate the data object
for date in df_J.date.unique():

    # I select the date (and remove DC for now)
    df_date = df_J[(df_J['state']!= 'District of Columbia' ) &  (df_J['date']== date )]

    for col in df_date.columns:  # I transform the columns into string type so I can:
        df_date[col] = df_date[col].astype(str)

    ### I create the text for mouse-hover for each state, for the current date    
    df_date['text'] = df_date['state'] + ' sentiment: ' + df_date['Value_Average_Past_7_days']

    ### create the dictionary with the data for the current year
    data_one_date = dict(
                        type='choropleth',
                        locations = df_date['State_code'],
                        z=df_date['Value_Average_Past_7_days'].astype(float),
                        locationmode='USA-states',
                        colorscale = scl,
                        text = df_date['text'],
                        colorbar = dict(
                        title = "Average 7 day sentiment")
                        )
    data_slider.append(data_one_date)  # I add the dictionary to the list of dictionaries for the slider

##  I create the steps for the slider
steps = []
for i in range(len(data_slider)):
    step = dict(method='restyle',
                args=['visible', [False] * len(data_slider)],
                label='Date {}'.format(i+1)) # label to be displayed for each step (year)
    step['args'][1][i] = True
    steps.append(step)

##  I create the 'sliders' object from the 'steps' 
sliders = [dict(active=0, pad={"t": 1}, steps=steps)]  

# I set up the layout (including slider option)
layout = dict(title = 'Average 7-day Sentiment on Covid-19 Protection Measures in June',
              geo=dict(scope='usa',
              projection={'type': 'albers usa'}),
              sliders=sliders)

# I create the figure object:
fig = dict(data=data_slider, layout=layout) 

# to plot in the notebook
plotly.offline.iplot(fig)
```
